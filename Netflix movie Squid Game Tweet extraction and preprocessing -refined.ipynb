```python
import twint
import nest_asyncio
nest_asyncio.apply()
import pandas as pd
import os 
import time
import re

```

```python
UK=["Essex","Newcastle","London","Portsmouth","Leicestershire","Kent","Cardiff","Glasgow"]
USA=["New York","Los Angeles","Chicago","Phoenix","Philadelphia","Dallas","San Diego","San Jose","San Antonio","Calif"]
Europe=["Paris","Lisbon","Amsterdam","Florence","Vienna","Berlin","Dublin"]
Others=["Hong Kong","Maldives","Mubai","Seoul","Manila","Lagos","Abuja","Cape town"]
all_countries=UK+USA+Europe+Others
print(all_countries)
#Countries = ["Cardiff","Newcastle","London","Manchester","Lagos","Abuja","Cape town","Seoul","Berlin","New Delhi","New York","California","Manila"]
ldf = list()
for city in all_countries:
    print(city)
    c = twint.Config()
    c.Search = "Squid Game"
    c.Lang= "en"
    c.Pandas = True
    c.Since = "2021-12-12"
    c.Near = city
    c.Limit = 20000
    #c.Location=True
    twint.run.Search(c)
    tmp=twint.output.panda.Tweets_df[['date','username','tweet','language','hashtags','near']]
    ldf.append(tmp)
data = pd.concat(ldf, ignore_index = True)
data
    
```

```python
# Remove duplicate tweet
data1=data.drop_duplicates(subset=['date','tweet'],keep='first')
data1.shape
```

```python
# Function to clean the data
def clean_text(text):  
    pat1 = r'@[^ ]+'                   
    pat2 = r'https?://[A-Za-z0-9./]+'  
    pat3 = r'\'s'                      
    pat4 = r'\#\w+'                     
    pat5 = r'&amp '                     
    pat6 = r'[^A-Za-z\s]'               
    combined_pat = r'|'.join((pat1, pat2,pat3,pat4,pat5, pat6))
    text = re.sub(combined_pat,"",text).lower()
    return text.strip()
```

```python
data1["cleaned_tweet"] = data1["tweet"].apply(clean_text)
```

```python
data1
```

```python
# Sentiment Scoring
from textblob import TextBlob
#word cloud
from wordcloud import WordCloud
```

```python
# Sentiment Analysis
print("Running sentiment process")
for row in data1.itertuples():
    tweet = data1.at[row[0], 'cleaned_tweet']
#run sentiment using TextBlob
    analysis = TextBlob(tweet)
#set value to dataframe
    data1.at[row[0], 'polarity'] = analysis.sentiment[0]
    data1.at[row[0], 'subjectivity'] = analysis.sentiment[1]
#Create Positive / negative column depending on polarity
    if analysis.sentiment[0]>0:
        data1.at[row[0], 'Sentiment'] = "Positive"
    elif analysis.sentiment[0]<0:
        data1.at[row[0], 'Sentiment'] = "Negative"
    else:
        data1.at[row[0], 'Sentiment'] = "Neutral"
        
```
